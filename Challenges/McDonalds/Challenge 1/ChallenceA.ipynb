```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

x = pd.read_csv('menu.csv',sep=',')
df = pd.DataFrame(x)

#print(df);
#df.describe(include = "all")
# untuk nampilin rata" semua katagoris kalori
#tampil=df[['Calories']].describe()
#print(tampil)

#soal A
df_gptest = df[['Category','Calories']]
grouped_test_Category = df_gptest.groupby(['Category'],as_index= True).mean()
grouped_test_Category
print(grouped_test_Category)

grouped_test_Category.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Categories') # add to x-label to the plot
plt.ylabel('Average') # add y-label to the plot
plt.title('MCD Categories from Calories') # add title to the plot

plt.show()
```
